In the early twentieth century, ciphering systems were widely in use as they enabled the Western Ally countries during World War II to read and understand substantial amounts of Morse-coded radio communications of the Axis powers. However, even after multiple efforts, these systems failed to break the codes that were enciphered using the Nazi encryption machine Enigma. After years of trial and error, Polish mathematician Marian Rejewski broke the code of Enigma using only limited documentation obtained by French military intelligence.
<br>
<br>
Less than a decade after breaking the Enigma, British computer scientist Alan Turing challenged the world with one of the simplest yet perplexing questions of all time - “Can machines think?” In 1950, he published his paper ‘Computing Machinery and Intelligence’ and proposed the Turing Test which he fondly called an ‘imitation game’ which tests whether a computer could fool humans into thinking that they were communicating with another human being. This established a strong base for the vision of artificial intelligence(AI) - an advancement that allows systems to mimic or even outperform human cognitive abilities.
<br>
<br>
Since then, the field of AI has been seeing a constant boom and this growth has been enabled by exponentially faster computers and large datasets. Today, AI forms the basis of the world's popular applications like voice recognition, self-driven automobiles, controlling robots, and android assistants.

```python
from cProfile import label
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("data\papers.csv")
# df=pd.read_csv(r"C:\Users\jainp\OneDrive\Desktop\writing_assign\writing-assignment-shridhar_likhita_paras\_notebooks\data\papers.csv")
```

```python
sorted_yr=df.sort_values("Year")
remove_dupli=sorted_yr.drop_duplicates(subset='Title', keep="last")
d = dict(remove_dupli["Year"].value_counts())
D = dict(reversed(list(d.items())))
plt.figure(figsize=(25,10))
x = [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
y=np.zeros(len(x))
for i in range(len(x)):
    y[i]=D[x[i]]
# plt.plot(x,list(D.values()))
# plt.bar(x, list(D.values()))
plt.bar(x,y)
# plt.xticks(x, list(D.keys()))
plt.title("Plot 1: Number of AI publications per year")
plt.show()
```

The total number of papers published at conferences increased from 2006 to 2020, as shown in Plot 1. In 2008, the number was around 250, and by 2020, it would have risen to 3000. It's a significant shift, indicating a growing interest in AI research. The disparity has widened in recent years compared to the previous decade. The reason could be the introduction of AI in more conferences and the novelty in the resources.  It is indeed linked to more powerful computing processors and the availability of massive dataset storage. Additionally, the rising usage of the internet and automation has sparked interest in the computer science department and its various disciplines. AI is one of the leading fields that have witnessed an increased number of paper publications and conferences. The decrease in the number of papers published in the year 2021 is due to the fact that the dataset does not include all the papers published in 2021. It only has those that are published in ICLR conference.

```python
data=df.drop_duplicates(['Title','Affiliation'])
c = dict(data["Affiliation"].value_counts(ascending=False))
C = dict(list(c.items())[0: 10])
plt.figure(figsize=(25,10))
plt.bar(range(len(C)), list(C.values()))
plt.xticks(range(len(C)), list(C.keys()))
plt.title("Plot 2: Total number of AI publications by the top 10 institutions")
plt.show()
```

AI is on the upswing all across the globe. Some world institutions are well ahead of others in terms of research and publications. Plot 2 depicts the top five institutions that produced the most publications between the years 2006 and 2021. Google is a worldwide tech business that has published over 700 papers in the last few years and is at the top. Universities account for four of the top five organizations, all of which are based in the United States. The plot clearly demonstrates America's superiority in AI.
<br>
<br>
Over the last two decades, the United States has been the leader in AI-related research output, with the highest number of publications. The main academic institutions in the field of AI include 'Stanford,' 'Carnegie Mellon,' 'UC Berkley,' and 'MIT.' One thing to keep in mind is that, while these universities are located in the United States, the Ph.D. students conducting research there come from all over the world. This element has the potential to alter the entire interpretation of leading AI countries. Furthermore, all of these institutions have substantial earnings and massive funding. These elements are critical for AI research and development.

```python
n = 10
a = data['Affiliation'].value_counts()[:n].index.tolist()
x = [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]

data_google=data.loc[df['Affiliation'] == "Google"]
d_google = dict(data_google["Year"].value_counts())
D_google = dict(reversed(list(d_google.items())))
y_google= list(D_google.values())
# plt.scatter(range(len(D_google)), list(D_google.values()),color='maroon')
# plt.xticks(range(len(D_google)), list(D_google.keys()))
# plt.show()

data_stan=data.loc[df['Affiliation'] == "Stanford University"]
d_stan = dict(data_stan["Year"].value_counts())
D_stan = dict(reversed(list(d_stan.items())))
y_stan = list(D_stan.values())
# plt.scatter(range(len(D_stan)), list(D_stan.values()),color='green')
# plt.xticks(range(len(D_stan)), list(D_stan.keys()))

data_carnegie = data.loc[df['Affiliation'] == "Carnegie Mellon University"]
d_carnegie = dict(data_carnegie["Year"].value_counts())
D_carnegie = dict(reversed(list(d_stan.items())))
y_carnegie = list(D_carnegie.values())

data_ucberkeley = data.loc[df['Affiliation'] == "UC Berkeley"]
d_ucberkeley = dict(data_ucberkeley["Year"].value_counts())
D_ucberkeley = dict(reversed(list(d_ucberkeley.items())))
y_ucberkeley = list(D_ucberkeley.values())

data_mit=data.loc[df['Affiliation'] == "MIT"]
d_mit = dict(data_mit["Year"].value_counts())
D_mit = dict(reversed(list(d_mit.items())))
y_mit=list(D_mit.values())
plt.plot(x,y_google,'b',x,y_stan,'r',x,y_carnegie,'y',x,y_ucberkeley,'c',x,y_mit,'m')
plt.title("Plot 3: Number of publications by top 5 institutions throughtout the years")
plt.legend(['Google', 'Stanford University', 'Carnegie Mellon University', 'UC Berkeley', 'MIT'])
plt.show()
```

We acquired Plot 3 by focusing on the top 5 significant institutes in the field of AI and plotting their number of publications year over year. The story demonstrates that these organizations have been competing for the top spot. Despite the fact that all of these universities have boosted their publications, some have outperformed others. For example, in the early years, 'Google' was ranked behind 'UC Berkeley' but has since exceeded all in recent years. In 2006, 'MIT' was ranked fifth; however, by 2020, it had risen to third place. The graph shows that even if an institution is below the others at a point in time, there is a fair chance for it to climb up the chart. This is indicated by the zig-zag pattern in the plot's midsection. It demonstrates their competition to stay ahead of the competition. University competitions have accelerated AI's growth and yielded spectacular outcomes thus far.

<br>
From the above-mentioned observations, one thing is pretty clear that the growth rate of AI in the US is immensely higher than that of Europe. The primary reason behind this is that the United States has access to large data sets for training machine-learning algorithms. Europe, on the other hand, has strict data protection rules that safeguard people's privacy. These strict data protection laws limit the number of resources available to train AI systems leading to uncertainty on whether Europe will ever be able to compete with the US in terms of the growth of AI by generating very complex algorithms.
The application of AI systems' knowledge to actual problems, rather than academic concerns, drives the improvement of AI systems. Countries will most likely achieve progress by deploying AI in domains where they currently excel.
