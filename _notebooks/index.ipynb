In the early twentieth century, ciphering systems were widely in use as they enabled the Western Ally countries during World War II to read and understand substantial amounts of Morse-coded radio communications of the Axis powers. However, even after multiple efforts, these systems failed to break the codes that were enciphered using the Nazi encryption machine Enigma. After years of trial and error, Polish mathematician Marian Rejewski broke the code of Enigma using only limited documentation obtained by French military intelligence.
<br>
Less than a decade after breaking the Enigma, British computer scientist Alan Turing challenged the world with one of the simplest yet perplexing questions of all time - “Can machines think?” In 1950, he published his paper ‘Computing Machinery and Intelligence’ and proposed the Turing Test which he fondly called an ‘imitation game’ which tests whether a computer could fool humans into thinking that they were communicating with another human being. This established a strong base for the vision of artificial intelligence(AI) - an advancement that allows systems to mimic or even outperform human cognitive abilities.

```python
from cProfile import label
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("data\papers.csv")
# df=pd.read_csv(r"C:\Users\jainp\OneDrive\Desktop\writing_assign\writing-assignment-shridhar_likhita_paras\_notebooks\data\papers.csv")
```

```python
sorted_yr=df.sort_values("Year")
remove_dupli=sorted_yr.drop_duplicates(subset='Title', keep="last")
d = dict(remove_dupli["Year"].value_counts())
D = dict(reversed(list(d.items())))
plt.figure(figsize=(25,10))
x = [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
y=np.zeros(len(x))
for i in range(len(x)):
    y[i]=D[x[i]]
# plt.plot(x,list(D.values()))
# plt.bar(x, list(D.values()))
plt.bar(x,y)
# plt.xticks(x, list(D.keys()))
plt.title("Plot 1: Number of AI publications per year")
plt.show()
```

```python
data=df.drop_duplicates(['Title','Affiliation'])
c = dict(data["Affiliation"].value_counts(ascending=False))
C = dict(list(c.items())[0: 10])
plt.figure(figsize=(25,10))
plt.bar(range(len(C)), list(C.values()))
plt.xticks(range(len(C)), list(C.keys()))
plt.title("Plot 2: Total number of AI publications by the top 10 institutions")
plt.show()
```

```python
n = 10
a = data['Affiliation'].value_counts()[:n].index.tolist()
x = [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]

data_google=data.loc[df['Affiliation'] == "Google"]
d_google = dict(data_google["Year"].value_counts())
D_google = dict(reversed(list(d_google.items())))
y_google= list(D_google.values())
# plt.scatter(range(len(D_google)), list(D_google.values()),color='maroon')
# plt.xticks(range(len(D_google)), list(D_google.keys()))
# plt.show()

data_stan=data.loc[df['Affiliation'] == "Stanford University"]
d_stan = dict(data_stan["Year"].value_counts())
D_stan = dict(reversed(list(d_stan.items())))
y_stan = list(D_stan.values())
# plt.scatter(range(len(D_stan)), list(D_stan.values()),color='green')
# plt.xticks(range(len(D_stan)), list(D_stan.keys()))

data_carnegie = data.loc[df['Affiliation'] == "Carnegie Mellon University"]
d_carnegie = dict(data_carnegie["Year"].value_counts())
D_carnegie = dict(reversed(list(d_stan.items())))
y_carnegie = list(D_carnegie.values())

data_ucberkeley = data.loc[df['Affiliation'] == "UC Berkeley"]
d_ucberkeley = dict(data_ucberkeley["Year"].value_counts())
D_ucberkeley = dict(reversed(list(d_ucberkeley.items())))
y_ucberkeley = list(D_ucberkeley.values())

data_mit=data.loc[df['Affiliation'] == "MIT"]
d_mit = dict(data_mit["Year"].value_counts())
D_mit = dict(reversed(list(d_mit.items())))
y_mit=list(D_mit.values())
plt.plot(x,y_google,'b',x,y_stan,'r',x,y_carnegie,'y',x,y_ucberkeley,'c',x,y_mit,'m')
plt.title("Plot 3: Number of publications by top 5 institutions throughtout the years")
plt.legend(['Google', 'Stanford University', 'Carnegie Mellon University', 'UC Berkeley', 'MIT'])
plt.show()
```
